```r
install.packages("readxl")
library(readxl)
```

```r
datos<- read_excel("data.xlsx")
head(datos)
```

```r
mean(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]), 2,mean)
```

```r
lapply(datos[,2:4], mean)
```

```r
unlist(lapply(datos[,2:4], mean)) # unlist -> para no verlo como lista
```

```r
median(datos$Estatura)
```

```r
apply(as.matrix(datos[,2:4]),2, median)
```

```r
unlist(lapply(datos[,2:4], median)) # unlist -> para no verlo como lista
```

```r
var(datos$Estatura)
```

```r
sd(datos$peso)
```

```r
apply(as.matrix(datos[,2:4]), 2, sd)
```

```r
unlist(lapply(datos[,2:4], sd)) # unlist -> para no verlo como lista
```

```r
quantile(datos$Estatura)
```

Esta práctica muestra cómo obtener e interpretar, en R, los principales estadísticos descriptivos de un conjunto de datos para resumir su comportamiento sin necesidad de revisar registro por registro. Se aplicaron medidas de tendencia central (media y mediana) y dispersión (varianza y desviación estándar), además de medidas de localización (cuantiles y percentiles), lo que permite entender rápidamente cuál es el valor “típico” de las variables y qué tan dispersos están los datos.
